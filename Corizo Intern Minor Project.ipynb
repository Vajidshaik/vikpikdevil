# Clustering - K Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
plt.scatter(v1["track_popularity"],v1["danceability"])
```

```python
#track_popularity	danceability	energy	key	
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=5)
km.fit(v1[["track_popularity","danceability"]])
```

```python
km.cluster_centers_
```

```python
v1["cluster_group"]=km.labels_
v1
```

```python
sns.scatterplot(x="track_popularity",y="danceability",hue="cluster_group",data=v1)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ip = v1.select_dtypes(include=['number', 'bool'])
op = v1['playlist_genre']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(ip, op, train_size=0.8, random_state=42)
```

```python
v3 = RandomForestClassifier(n_estimators=100, random_state=42)
v3.fit(xtrain, ytrain)
```

```python

predictions = v3.predict(xtest)
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ytest, predictions)
```

```python
print(f"Accuracy: {accuracy:.2%}")
```
